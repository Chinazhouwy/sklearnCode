```python
import pandas as pd
data = pd.read_csv('data.csv')
print(data.head())
```

```python
x = data.loc[:,'x']
y = data.loc[:,'y']
print(x,y)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(5,5))
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
print(x,y)
```

```python
import numpy as np
x = np.array(x)
# print(type(x),x.shape)
x = x.reshape(-1,1)
y = np.array(y)
# print(type(x),x.shape)
y = y.reshape(-1,1)
print(x)
```

```python
lr_model.fit(x,y)
```

```python
y_predict = lr_model.predict(x)
print(y_predict)
```

```python
y_predict1 = lr_model.predict(np.array(100).reshape(-1,1))
print(y_predict1)
```

```python
y_predict2 = lr_model.predict([[3.5]])
print(y_predict2)
```

```python
# 斜率
a = lr_model.coef_
# 截距
b = lr_model.intercept_
print(a,b)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print(y_predict)
MSE = mean_squared_error(y,y_predict)
R2 = r2_score(y,y_predict)
print(MSE,R2)
```

```python
plt.figure()
plt.plot(y,y_predict)
```

```python
data = pd.read_csv('usa_housing_price.csv')
data.head()
```

```python
fig = plt.figure(figsize=(10,10))
fig1 = plt.subplot(231)
plt.scatter(data.loc[:,'Avg. Area Income'],data.loc[:,'Price'])
plt.title('Price vs Income')
fig2 = plt.subplot(232)
plt.scatter(data.loc[:,'Avg. Area House Age'],data.loc[:,'Price'])
plt.title('Price vs House Age')
fig3 = plt.subplot(233)
plt.scatter(data.loc[:,'Avg. Area Number of Rooms'],data.loc[:,'Price'])
plt.title('Price vs Number of Rooms')
fig4 = plt.subplot(234)
plt.scatter(data.loc[:,'Area Population'],data.loc[:,'Price'])
plt.title('Price vs Population')
fig5 = plt.subplot(235)
plt.scatter(data.loc[:,'size'],data.loc[:,'Price'])
plt.title('Price vs size')
```

```python
x = data.loc[:,'size']
y = data.loc[:,'Price']
y.head()
```

```python
x = np.array(x).reshape(-1,1)
print(x)
```

```python
from sklearn.linear_model import LinearRegression
LR1 = LinearRegression()
LR1.fit(x,y)
```

```python
y_pred_1 = LR1.predict(x)
print(y_pred_1)
```

```python
## “简单的来说就是axis=0代表往跨行(down),而axis=1代表跨列(across) 
## 换句话说: 使用0值表示沿着每一列或行标签/索引值向下执行方法 使用1值表示沿着每一行或者列标签横向执行对应的方法”
x_multi = data.drop(['Price'],axis=1)
print(x_multi)
```

```python
LR_multi = LinearRegression()
LR_multi.fit(x_multi,y)
```

```python
y_predict_multi = LR_multi.predict(x_multi)
print(y_predict_multi)
```

```python
MSE_multi = mean_squared_error(y,y_predict_multi)
R2_multi = r2_score(y,y_predict_multi)
print(MSE_multi,R2_multi)
```

```python
# 斜率
a = LR_multi.coef_
# 截距
b = LR_multi.intercept_
print(a,b)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(y,y_predict_multi)
plt.show()
```
