```python
import pandas as pd
import numpy as np
data = pd.read_csv('data.csv')
data.head()
```

```python
X = data.drop(['labels'],axis=1)
y = data.loc[:,'labels']
X.head()
y.head()
```

```python
pd.value_counts(y)
```

```python
from matplotlib import pyplot as plt
fig1 = plt.figure()
plt.scatter(X.loc[:,'V1'],X.loc[:,'V2'])
plt.title('un-label')
plt.xlabel('V1')
plt.ylabel('V2')
plt.show()
```

```python
fig2 = plt.figure()
label0 = plt.scatter(X.loc[:,'V1'][y==0],X.loc[:,'V2'][y==0])
label1 = plt.scatter(X.loc[:,'V1'][y==1],X.loc[:,'V2'][y==1])
label2 = plt.scatter(X.loc[:,'V1'][y==2],X.loc[:,'V2'][y==2])

plt.title('un-label')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
plt.show()
```

```python
from sklearn.cluster import KMeans
KM = KMeans(n_clusters=3,random_state=0)
KM.fit(X)
```

```python
centers = KM.cluster_centers_
print(centers)
```

```python
fig3 = plt.figure()
plt.scatter(centers[:,0],centers[:,1])
plt.show()
```

```python
fig4 = plt.figure()
label0 = plt.scatter(X.loc[:,'V1'][y==0],X.loc[:,'V2'][y==0])
label1 = plt.scatter(X.loc[:,'V1'][y==1],X.loc[:,'V2'][y==1])
label2 = plt.scatter(X.loc[:,'V1'][y==2],X.loc[:,'V2'][y==2])
plt.scatter(centers[:,0],centers[:,1])

plt.title('un-label')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
plt.show()
```

```python
#V1 = 80 ,V2 = 60
y_predict_test = KM.predict([[80,60]])
print(y_predict_test)
```

```python
y_predict_test = KM.predict(X)
print(pd.value_counts(y_predict_test),pd.value_counts(y))
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict_test)
print(accuracy)
```

```python
y_correct = []
for i in y_predict_test:
    if i == 0:
        y_correct.append(1)
    elif i == 1:
        y_correct.append(2)
    else:
        y_correct.append(0)
```

```python
print(pd.value_counts(y_correct),pd.value_counts(y))
```

```python
accuracy = accuracy_score(y,y_correct)
print(accuracy)
```
